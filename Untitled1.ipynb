```python
import cv2
import mediapipe as mp
import numpy as np
import csv
import os
import time

# MediaPipe Hands
mp_hands = mp.solutions.hands
hands = mp_hands.Hands(static_image_mode=False, max_num_hands=1,
                       min_detection_confidence=0.7, min_tracking_confidence=0.7)
mp_drawing = mp.solutions.drawing_utils

# Output CSV file
csv_file = "landmarks.csv"

# If file doesn't exist, write header
if not os.path.exists(csv_file):
    with open(csv_file, "w", newline="") as f:
        writer = csv.writer(f)
        header = [f"x{i}" for i in range(63)] + ["label"]
        writer.writerow(header)

# Ask user for the label
label = input("Enter the label for this recording (e.g., A, B, Hello): ")

cap = cv2.VideoCapture(0)
print("Camera on. Starting capture in 5 seconds...")
time.sleep(5)

frames_to_capture = 20
captured_count = 0

while True:
    ret, frame = cap.read()
    if not ret:
        break

    frame = cv2.flip(frame, 1)
    rgb = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
    result = hands.process(rgb)

    if result.multi_hand_landmarks and captured_count < frames_to_capture:
        lm = result.multi_hand_landmarks[0].landmark
        xs = [l.x for l in lm]
        ys = [l.y for l in lm]
        zs = [l.z for l in lm]

        # Normalize
        minx, miny = min(xs), min(ys)
        row = []
        for i in range(21):
            row.append(xs[i] - minx)
            row.append(ys[i] - miny)
            row.append(zs[i])

        # Draw hand landmarks
        mp_drawing.draw_landmarks(frame, result.multi_hand_landmarks[0],
                                  mp_hands.HAND_CONNECTIONS)

        # Save frame automatically
        with open(csv_file, "a", newline="") as f:
            writer = csv.writer(f)
            writer.writerow(row + [label])
        captured_count += 1
        print(f"Captured frame {captured_count}/{frames_to_capture}")

    cv2.putText(frame, f"Label: {label}", (10, 40),
                cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 255, 0), 2)
    cv2.imshow("Collecting Data", frame)

    if captured_count >= frames_to_capture:
        print("Captured all frames!")
        break

    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
from tensorflow.keras.utils import to_categorical
import json

# Load dataset
data = pd.read_csv("landmarks.csv")

# Force labels to string and drop NaN/empty
data = data.dropna(subset=["label"])
data["label"] = data["label"].astype(str)

# Separate features and labels
X = data.drop("label", axis=1).values
y = data["label"].values

# Map string labels to integers
classes = sorted(list(set(y)))
class_to_idx = {c: i for i, c in enumerate(classes)}
y_num = np.array([class_to_idx[label] for label in y])

# One-hot encode labels for training
y_cat = to_categorical(y_num, num_classes=len(classes))

# Train-test split
X_train, X_test, y_train, y_test = train_test_split(
    X, y_cat, test_size=0.2, random_state=42, stratify=y_cat
)

# Build model
model = Sequential([
    Dense(256, activation="relu", input_shape=(X.shape[1],)),
    Dropout(0.3),
    Dense(128, activation="relu"),
    Dropout(0.3),
    Dense(64, activation="relu"),
    Dense(len(classes), activation="softmax")
])

model.compile(optimizer="adam", loss="categorical_crossentropy", metrics=["accuracy"])

# Train
history = model.fit(X_train, y_train, validation_data=(X_test, y_test), epochs=50, batch_size=32)

# Save model and classes
model.save("sign_language_model.h5")
with open("classes.json", "w") as f:
    json.dump(classes, f)

print("âœ… Model trained and saved as sign_language_model.h5")
```

```python
# Evaluate on test set
loss, accuracy = model.evaluate(X_test, y_test, verbose=0)
print(f"âœ… Test Accuracy: {accuracy * 100:.2f}%")
```

```python
import cv2
import mediapipe as mp
import tensorflow as tf
import pyttsx3
import pandas as pd
import numpy as np
import json

# Initialize the text-to-speech engine
engine = pyttsx3.init()

# Load the trained model and class labels
model = tf.keras.models.load_model('sign_language_model.h5')
with open('classes.json', 'r') as f:
    classes = json.load(f)

# Initialize MediaPipe Hands
mp_hands = mp.solutions.hands
mp_drawing = mp.solutions.drawing_utils
hands = mp_hands.Hands(static_image_mode=False, max_num_hands=1, min_detection_confidence=0.7)

# Start webcam capture
cap = cv2.VideoCapture(0)
last_prediction = ""

while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break
    
    # Flip the frame for a mirrored view
    frame = cv2.flip(frame, 1)
    
    # Convert the frame to RGB for MediaPipe
    rgb_frame = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
    
    # Process the frame to detect hands
    result = hands.process(rgb_frame)

    if result.multi_hand_landmarks:
        # Extract landmarks and flatten them
        landmarks = []
        for landmark in result.multi_hand_landmarks[0].landmark:
            landmarks.extend([landmark.x, landmark.y, landmark.z])
        
        # Make a prediction
        prediction = classes[np.argmax(model.predict(np.array([landmarks]))[0])]
        
        # Show on screen
        cv2.putText(frame, f"Prediction: {prediction}", (10, 40),
                    cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 255, 0), 2)
        mp_drawing.draw_landmarks(frame, result.multi_hand_landmarks[0],
                                  mp_hands.HAND_CONNECTIONS)

        # Speak prediction only if it changes
        if prediction != last_prediction:
            print(f"ğŸ—£ {prediction}")
            engine.say(prediction)
            engine.runAndWait()
            last_prediction = prediction

    cv2.imshow("Sign Language Recognition", frame)

    if cv2.waitKey(1) & 0xFF == ord("q"):
        break

cap.release()
cv2.destroyAllWindows()
```
